Q2

Split the labels DataFrame into two: labels on gold and labels unknown, the former containing all rows where the url is present in the gold set, and the latter one contains all remaining rows of labels.


```python
mask2 = labels_df['url'].isin(gold_df['url'])
#make a mask to find matching urls between dataframes
```

```python
labels_on_gold = labels_df[mask2]
#apply mask to labels to get a dataframe of matching urls 
```

```python
labels_on_gold
#check
```

```python
labels_unknown = labels_df[~mask2].drop_duplicates()
#apply opposite of mask to labels to get dataframe of remaining rows
```

```python
labels_unknown
#check
```

Q3

Create a rater goodness DataFrame that is indexed by turk, and has two columns: the number of ratings, and the average correctness of ratings for each turk (both on gold set urls).


```python
ratings_compared = gold_df.merge(labels_df,left_on='url',right_on='url',suffixes=['_gold', '_label'])
ratings_compared
#merging dataframes by matching urls, same as labels_on_gold except both category ratings are shown 
```

```python
ratings_compared['correct'] = ratings_compared['category_gold'] == ratings_compared['category_label']
ratings_compared
#adding a column that shows whether the turk was correct in its rating of a url 
```

```python
ratings_compared['correct']=ratings_compared[['correct']].astype(int)
#converting Booleans to 1's and 0's in correct column 
```

```python
ratings_compared
#check
```

```python
rater_goodness = ratings_compared.groupby(['turk'])['correct'].agg(['mean', 'count'])
#create groupby w/ mean as average correctness & count as number of ratings per turk 
```

```python
rater_goodness
#check 
```

Q4

If someone is correct p fraction of the time, the odds of success are defined
    as: (p/(1.001-p))
Attach a column called odds to the rater goodness DataFrame, using the
average correctness of the turk as his or her p.

```python
rater_goodness['odds'] = rater_goodness['mean']/(1.001-(rater_goodness['mean']))
#create formula for odds with mean as p, add column to datafram that displays odds calculated from this formula
```

```python
rater_goodness
#check
```

Q5

List the top 10 most accurate turks who have rated at least 20 gold set URLs.


```python
at_least_21 = rater_goodness[rater_goodness['count'] >= 20]
#get turks that have rated at least 20 gold set urls by applying a mask 
```

Q7

Consider each url u that is not in the gold set, and each category c. For the pair (u, c), calculate the product of odds of all turks who (a) rated url u as category c, and (b) have rated more gold set urls than 75% of all turks.
We shall call such products of odds the overall odds henceforth.

```python
number_turks = labels_on_gold['turk'].value_counts()
#get # of ratings in gold set for each turk 
turks_quantile = number_turks.quantile(.75)
#getting 75% of turks 
rater_goodness2 = rater_goodness[number_turks > turks_quantile]
#getting turks that have rated more than 75% by applying a mask 
```

```python
rater_goodness2['turk'] = rater_goodness2.index
#make column for turks
```

```python
rater_goodness2=rater_goodness2.reset_index(drop=True)
rater_goodness2
#reset index 
```

```python
odds_df = rater_goodness2.merge(labels_unknown, left_on = 'turk', right_on = 'turk', suffixes=['_rater', '_labels'], how='left')
odds_df 
#merge labels unknown with masked rater goodness dataframe 
```

```python
odds_group = odds_df.groupby(['url', 'category'])['odds'].prod().unstack()
#group the data by url, showing different ratings for each url and calculated odds
odds_group
```

Q8

Create a DataFrame (called result 75, whose index is URLs not in the gold set, and with two columns called top category and top odds. The top category should be the category with the highest overall odds for that url, and the top odds should be the overall odds for that top category.
These are our predictions, and the confidence we have in them (higher overall odds implies greater confidence). If you want, you can check to see if the predicted categories make sense.

```python
#transpose and unstack odds_group dataframe from #7 that shows the rating for the url from the top turks and the odds for that category
result_75 = pd.DataFrame({'top category 75': odds_group.T.idxmax().values, 'top odds': odds_group.T.max().values}, index=odds_group.T.idxmax().index)
result_75
```
